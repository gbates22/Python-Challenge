```python
zip_charlotte=zip_csv.loc[zip_csv['ZipName']== 'Charlotte, NC']
zip_denver = zip_csv.loc[zip_csv['ZipName']== 'Denver, CO']
zip_charlotte
#zip_denver
```

```python
# charlotte_zip_sort= zip_charlotte.sort_values(['Median Listing Price'], ascending=False)
# print(charlotte_zip_sort)

charlotte = zip_charlotte[['ZipCode','Month', 'Median Listing Price', 'Avg Listing Price', 'Total Listing Count']]
charlote = charlotte.to_string(columns = ['Month'])
charlote= charlotte.to_string(columns = ['ZipCode']) 

charlotte = charlotte.groupby(charlotte['ZipCode'], as_index=True)
charlotte_ALP = charlotte['Avg Listing Price'].mean()
charlotte_MLP = charlotte['Median Listing Price'].mean()
charlotte_TLC = charlotte['Total Listing Count'].mean()
charlotte_ALP= pd.DataFrame(charlotte_ALP)

charlotte_ALP['Avg Listing Price']= pd.to_numeric(charlotte_ALP['Avg Listing Price'])
charlotte_ALP.plot(kind='bar', use_index=True, title='Average Listing Price in Charlotte NC by Zipcode 2011-2017', figsize=(15,4))


x_axis = np.arange(len(charlotte_ALP))
tick_loc = [value for value in x_axis]
plt.xticks(tick_loc, rotation=45)


charlotte_ALP['Avg Listing Price']=charlotte_ALP['Avg Listing Price'].map('${:,.2f}'.format)
print(charlotte_ALP.dtypes)

plt.tight_layout()
plt.savefig('charlotte_ALP')
```
